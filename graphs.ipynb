```python
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

plt.rcParams.update({'font.size': 16})
```

```python
def plot_experiments(results, labels, x=None, n_trials=100, confidence=0.95, lim=(0, 1), xlabel=''):
  plt.figure(figsize=(8, 5))
  for result, label, color in zip(results, labels, mcolors.TABLEAU_COLORS):
    if x is None:
      x = np.arange(result.shape[0])+1
    se = np.sqrt(result * (100 - result) / n_trials)
    z_score = norm.ppf((1+confidence)/2)
    ci = z_score * se

    plt.errorbar(np.arange(x.shape[0]), result, yerr=ci, fmt='o', capsize=5, label=label, color=color)
    plt.plot(np.arange(x.shape[0]), result, linestyle='-', color=color)
    plt.xticks(ticks=np.arange(x.shape[0]), labels=x)
    plt.xlabel(xlabel)
    plt.ylabel('Win Rate')
    plt.ylim(*lim)
  plt.legend()
  plt.show()
```

```python
plot_experiments([np.array([42, 40, 26.5, 33])], ['MCTS-MR'], xlabel='Minimax Depth',  lim=(0, 55))
```

```python
plot_experiments([np.array([57.5, 46.5, 48, 49, 49.5, 42])], ['MCTS-MR'], xlabel='Minimax Depth',  lim=(30, 70))
```

```python
plot_experiments(
    [np.array([29.5, 36, 41, 49, 41.5, 46.5, 60.5, 54.5, 53.5, 45, 44]),
     np.array([17, 23.5, 33.5, 46, 51.5, 56.5, 38, 52.5, 58, 53.5, 53.5])],
     ['MCTS-MS-2', 'MCTS-MS-4'],
    np.array([0, 1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]),
    xlabel='Visit Limit',  lim=(0, 75)
)
```
